```python
arr = np.array([[1, 2, 3, 4, 5, 0],
                [2, 4, 6, 8, 10, 12]])

diff = np.diff(arr, axis = 1)
print(diff)

unique, count = np.unique(diff, axis = 1, return_counts = True)
print(unique)
print(count)
```
